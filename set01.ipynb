```python
x = [10,20,13,12,15]
print(x)
```

```python
max = x[0]
for i in range(1,len(x),1):
    if x[i] > max:
      max=x[i]
      id = i
      print(max,i)
    else:continue
print(f"the maximum is {max} at index {id}")
```

```python
min = x[0]
for j in range(1,len(x),1):
    if x[j] < min:
        min = x[j]
        id = j
        print(min,j)
    else:continue
print(f"the minimum is {min} at index {id}")
```

```python
list = [10,12,32,31,34]
even_count=0
odd_count=0

for num in list:
    if num%2==0:
        even_count +=1
    else:
        odd_count +=1
print("Number of even numbers:", even_count)
print("Number of odd numbers:", odd_count)
```
